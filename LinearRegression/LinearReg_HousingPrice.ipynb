# Spliting data in train and test and Applying Linear Regression 

```python
import sklearn
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sclr = MinMaxScaler(feature_range=(0,1))
scld_data= sclr.fit_transform(features)
feacture_scl=pd.DataFrame(scld_data,columns=features.columns)
```

```python
feacture_scl
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(feacture_scl,Target, random_state=42)
#train_test_split(, test_size=0.33, randon_state=33)
```

```python
from sklearn.linear_model import LinearRegression
lnReg = LinearRegression()
lnReg.fit(x_train, y_train)
```

```python
y_pred = lnReg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
