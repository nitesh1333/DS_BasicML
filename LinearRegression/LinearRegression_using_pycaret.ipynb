```python
# Importing data using pandas
import pandas as pd
data = pd.read_csv('../Dataset/Linear_USA_Housing.csv')
```

```python
data.head()
```

```python
# dropping column Address as this column is not having any importance
data.drop(['Address'], axis=1,inplace=True)
```

```python
from pycaret.regression import *
lreg = setup(data = data, target = 'Price', normalize = True, remove_multicollinearity = True)
```

```python
# Create Linear Regression Model and train linear regression model
lr_model = create_model('lr') #lr is the id of the model
```

```python
lr_tuned = tune_model(lr_model, round=1)
```

```python
plot_model(lr_tuned)
```
